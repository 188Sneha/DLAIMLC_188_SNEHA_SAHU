```python
#Perceptron from scratch
x = [0.75, 0.65, 0.80]
w =[0.4, 0.3, 0.5]
b=-0.2
y=1
n=0.01
```

```python
z=x[0]*w[0]+x[1]*w[1]+x[2]*w[2]+b
print(z)

#activation fn
if(z>=0):
  y_pred=1
else:
  y_pred=0
print(y_pred)
```

```python
error=y-y_pred
while(error!=0):
  wnew=[]
  for i in range(0,len(w)):
    w[i]=w[i]+(n*error*x[i])

  bnew=b+(n*error)

  z=x[0]*w[0]+x[1]*w[1]+x[2]*w[2]+b

  if(z>=0):
    y_pred=1
  else:
    y_pred=0

  error=y-y_pred
```
